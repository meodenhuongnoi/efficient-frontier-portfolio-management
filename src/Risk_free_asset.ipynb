```python
import pandas as pd
```

```python
# Read the CSV data with a custom date parser
date_parser = lambda x: pd.to_datetime(x, dayfirst=True, infer_datetime_format=True)
df_daily = pd.read_csv('riskfree.csv', parse_dates=['Date'], date_parser=date_parser)
print("Loaded Daily Data:")
print(df_daily)

# Step 2: Set 'Date' as the DataFrame index
df_daily.set_index('Date', inplace=True)

# Step 3: Resample to Monthly Frequency
# We'll take the last available price of each month
df_monthly = df_daily.resample('M').last()
print("\nAggregated Monthly Data:")
print(df_monthly)

# Step 4: Calculate Monthly Returns
# Monthly return = (Price_end_of_month / Price_end_of_previous_month) - 1
df_monthly['Monthly_Return'] = df_monthly['Price'].pct_change()
print("\nMonthly Returns:")
print(df_monthly)

# Step 5: Calculate the Average Monthly Return
average_monthly_return = df_monthly['Monthly_Return'].mean()
print(f"\nAverage Monthly Return: {average_monthly_return:.4f} ({average_monthly_return * 100:.2f}%)")
```

```python
import pandas as pd 
import numpy as np
import matplotlib as mplt
import mplfinance as mplf
```

```python
dfHSG = pd.read_csv("Dữ liệu Lịch sử HSG.csv")
dfBVS = pd.read_csv("Dữ liệu Lịch sử BVS.csv")
dfDCM = pd.read_csv("Dữ liệu Lịch sử DCM.csv")
dfPNJ = pd.read_csv("Dữ liệu Lịch sử PNJ.csv")
dfVND = pd.read_csv("Dữ liệu Lịch sử VND.csv")
```

```python
df_list =[dfHSG, dfBVS, dfDCM, dfPNJ, dfVND]

for df in df_list:
    df.drop(['Mở', 'Cao', 'Thấp', 'KL', '% Thay đổi'], axis=1, inplace=True)
```

```python
heading = ['HSG', 'BVS', 'DCM', 'PNJ', 'VND']
for i, df in enumerate(df_list):
    df.rename(columns={'Lần cuối': heading[i]}, inplace=True)
```

```python
for df in df_list[1:]: 
    df.drop("Ngày",axis=1, inplace=True)
```

```python
df = pd.concat([dfHSG, dfBVS, dfDCM, dfPNJ, dfVND], axis=1)
df.dropna(inplace=True)
df
```

```python
df.iloc[:, 1:] = df.iloc[:, 1:].replace(',', '', regex=True).astype(float)
df.dtypes
df
```

```python
df.to_csv('Stock_daily_price.csv', index=False)
```
